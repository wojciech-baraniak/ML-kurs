```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
data = np.genfromtxt('DataGD.csv', delimiter=',', skip_header=1)
data= np.hsplit(data, [7])
X = data[0]
y = data[1].reshape(807)
normf = np.array([5, 12,256, 6, 64,20,5000])
normy = 8000
```

```python
X = X/normf
y = y/normy
```

```python
print(X)
print(y)
print(X.shape)
print(y.shape)
```

```python
def predict(X, w,b):
    return np.dot(X,w)+b
```

```python
def computeCost(X,y, w,b):
    cost = 0
    for i in range(len(X)):
        cost += (np.dot(X[i], w)+b - y[i])
    cost = cost*(1/(len(X)*2))
    return cost
```

```python
def computeGradient(x,y,w,b):
    dj_dw = np.zeros(X.shape[1])
    dj_db = 0

    for i in range(X.shape[0]):
        err = ( np.dot(x[i],w) + b) - y[i]
        for j in range(X.shape[1]):
            dj_dw[j] += err*X[i,j]
        dj_db += err
    dj_db /= X.shape[0]
    dj_dw /= X.shape[0]
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, num_iters): 
    w = copy.deepcopy(w_in)
    b = b_in
    for i in range(num_iters):
        dj_db, dj_dw = computeGradient(X, y, w, b)
        w -= alpha*dj_dw
        b -= alpha*dj_db
    return w,b 

```

```python
w = np.array([1., 1., 10., 0.5,0.2, 0.2, 0.05, ])
b = 0.
iterations = 10000
alpha = 5.0e-7
w,b = gradient_descent(X, y, w,b,alpha,iterations)
print(w,b)
```
